```python
# missing values

df.isnull().sum()
```

```python
# Since review rating is numerical

# handling missing values

df['Review Rating']= df['Review Rating'].fillna(df['Review Rating'].median())
```

```python
df['Review Rating'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
# we can see in the data--- the snakecases
# so all should be lowercase
## why to change? --- it make code easier tp read & consistence 

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# we need to know the age group for classification of shopping category
# we create age_group

# Why pd.cut()?? --- to create meaningful age segments aligned with business understanding instead of purely disturbution based binning

df['age_group']=pd.qcut(df['age'], q=4, labels=['Adult','Young Adult','Middle-aged','Senior'])
```

```python
df[['age','age_group']].head(-1)
```

```python
# creating purchase_frequency_days
freq_mapping={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']= df['frequency_of_purchases'].map(freq_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(40)
```

```python
df[['promo_code_used','discount_applied']].head(5)
```

```python
### discount_apply and promocode_used is commonly same...
## as if you use discount_apply obviously promocode is used 
##but sometimes companies might be use discounts without promocode like seasonal changes,exclusive discounts for subscribed ones!!
```

```python
(df['promo_code_used']==df['discount_applied']).all()
```

```python
## that means it remains of same information in both columns so we delete / drop promocode_used
df=df.drop('promo_code_used',axis=1)
```
